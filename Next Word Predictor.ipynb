```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([dataset])
```

```python
len(tokenizer.word_index)
```

```python
from bs4 import BeautifulSoup
import re
num=0
input_sentences=[]
for text in dataset.split('\n'):

    newString = text.lower()
    newString = BeautifulSoup(newString, "lxml").text
    newString = re.sub(r'\([^)]*\)', '', newString)
    newString = re.sub('"','', newString)
    newString = ' '.join([contraction_mapping[t] if t in contraction_mapping else t for t in newString.split(" ")])    
    newString = re.sub(r"'s\b","",newString)
    newString = re.sub("[^a-zA-Z]", " ", newString) 
    newString = re.sub('[m]{2,}', 'mm', newString)
    if(num==0):
        tokens = [w for w in newString.split() if not w in stop_words]
    else:
        tokens=newString.split()
    long_words=[]
    for i in tokens:
        if len(i)>1:                                                 #removing short word
            long_words.append(i)   
    sent=(" ".join(long_words)).strip()
    tokenized_sentence=tokenizer.texts_to_sequences([text])[0]
    for i in range(1,len(tokenized_sentence)):
        input_sentences.append(tokenized_sentence[:i+1])
```

```python
input_sentences
```

```python
max_len=max([len(x) for x in input_sentences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sentences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
padded_input_sequences.shape
```

```python
x=padded_input_sequences[:,:-1]
padded_input_sequences[:,:-1].shape
```

```python
y=padded_input_sequences[:,-1]
padded_input_sequences[:,-1].shape
```

```python
from tensorflow.keras.utils import to_categorical
print(max(y) + 1)
num_classes = max(y) + 1
y = to_categorical(y, num_classes=num_classes)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout
```

```python

model = Sequential()
model.add(Embedding(446, 100, input_length=321))
model.add(LSTM(50, return_sequences=True))  # Updated the LSTM output units to match Dense layer
model.add(Dropout(0.4))
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(50))
model.add(Dense(446, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=100)
```
